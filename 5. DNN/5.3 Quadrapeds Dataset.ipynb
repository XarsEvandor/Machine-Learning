# Create random Quadrapeds dataset 
It creates the dataset, normalizes the feature values, and splits into training and validation set.

```python
from datasets.quadrapeds import CQuadrapedsDataSet
from sklearn import preprocessing
from mllib.visualization import CPlot

oDataset = CQuadrapedsDataSet(1000)

TEST_SAMPLE = 399
print("Features of sample #%d:" % (TEST_SAMPLE+1), oDataset.Samples[TEST_SAMPLE])
oMinMaxScaler = preprocessing.MinMaxScaler().fit(oDataset.Samples)
oDataset.Samples = oMinMaxScaler.transform(oDataset.Samples)
print("Minmax normalized features of sample #%d:" % (TEST_SAMPLE+1), oDataset.Samples[TEST_SAMPLE])
print("Target label of sample #%d:" % (TEST_SAMPLE+1), oDataset.Labels[TEST_SAMPLE])
print("Class description of sample #%d:" % (TEST_SAMPLE+1), oDataset.ClassNames[oDataset.Labels[TEST_SAMPLE]])
print("."*80)
oDataset.Split(0.2)

```

# Visualizing the feature space
We can visualize only a combinations of 2 out of 72 samples.

```python
sColorScheme = ["darkseagreen", "royalblue", "crimson", "goldenrod"] # https://matplotlib.org/3.1.0/gallery/color/named_colors.html

# Plot the training set 
oPlot = CPlot("Dataset", oDataset.Samples[:,6:8], oDataset.Labels
              ,p_sLabelDescriptions=oDataset.ClassNames, p_sColors=sColorScheme
              ,p_sXLabel="Feature 6", p_sYLabel="Feature 7"
              )
oPlot.Show(p_bIsMinMaxScaled=False)
                
                
# Plot the validation set
oPlot = CPlot("Validation Set", oDataset.VSSamples[:,6:8], oDataset.VSLabels
              ,p_sLabelDescriptions=oDataset.ClassNames, p_sColors=sColorScheme
              ,p_sXLabel="Feature 6", p_sYLabel="Feature 7"
              )
oPlot.Show(p_bIsMinMaxScaled=False)
```

# Class Distribution
We can see how many samples per class exist in our dataset using a **histogram**. If there are significantly more samples in a class compared to other classes, then we have an **imbalanced dataset**. Quadrapeds is a rather **balanced dataset**.

```python
from mllib.visualization import CHistogramOfClasses

print("Dataset samples shape: %s" % str(oDataset.Samples.shape))
print("Dataset labels shape: %s" % str(oDataset.Labels.shape))
oHistogram = CHistogramOfClasses(oDataset.Labels, oDataset.ClassCount)
oHistogram.Show()

```
