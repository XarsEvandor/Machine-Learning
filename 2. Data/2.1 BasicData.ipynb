# Samples
* Example of a Python class with a constructor
* Example of Python lists (arrays/collections)
* Example of instantiating an object
* Example of for-each loop that increases an index

```python
#=================================================================
class CDataSet(object):
  # --------------------------------------------------------
  def __init__(self):
    self.Samples = []
  # ------------------------------------------------------
#=================================================================


oDataSet = CDataSet()
oDataSet.Samples = [["Alpha","Beta"],[3,4],[0.5,0.25]]    

for nSampleIndex, oSample in enumerate(oDataSet.Samples):
  print("Sample %d:" % (nSampleIndex + 1), oSample)
```

Numpy Arrays for Features


```python
import numpy as np

oDataSet = CDataSet()
oDataSet.Samples = np.zeros((3,2), np.float32)

oDataSet.Samples[0,:] = [1, 2]
oDataSet.Samples[1,:] = [3, 4]
oDataSet.Samples[2,:] = [0.5, 0.25]


for nSampleIndex, oSample in enumerate(oDataSet.Samples):
  print("Sample%d:" % (nSampleIndex + 1), oSample)
```

Basic Slicing

```python
print("-"*40, "Matrix of our dataset", "-"*40)
print(oDataSet.Samples)

print("-"*40, "Vector of 2nd feature for all samples", "-"*40)
print(oDataSet.Samples[::2,:])
```
