# Using Pandas to Load from CSV files
* Loading a CSV file as a pandas dataframe object
* Converting a pandas dataframe object to a numpy array

```python
import pandas as pd

oDataFrame = pd.read_csv("SomeRawData.csv", delimiter=";");
oData = oDataFrame.to_numpy();

# Slicing
oIDs = oData[:, 0]
oTargets = oData[:, 1]
oFeatures = oData[:, 2:]

print("-"*30, "Targets", "-"*30)
print(oTargets)
print("-"*30, "Features", "-"*30)
print(oFeatures)
print(oFeatures.dtype)    
```

# Converting Text Field to a Categorical Feature
* Python dictionaries


```python
oCategoryIDs = dict()

nNextID = 0
for nIndex,sTextValue in enumerate(oFeatures[:,2]):
  print(sTextValue)
  if not sTextValue in oCategoryIDs:
    oCategoryIDs[sTextValue] = nNextID
    nNextID += 1
  nID = oCategoryIDs[sTextValue]
  oFeatures[nIndex, 2] = nID

print("-"*30, "Targets", "-"*30)
print(oTargets)
print("-"*30, "Features", "-"*30)
print(oFeatures)
print(oFeatures.dtype)                           
```

We ensure that the features array has only 32bit real numbers, before we use it to train a model

```python
import numpy as np

oFeatures = oFeatures.astype(np.float32)
print("-"*30, "Features", "-"*30)
print(oFeatures)
print(oFeatures.dtype)    
```
