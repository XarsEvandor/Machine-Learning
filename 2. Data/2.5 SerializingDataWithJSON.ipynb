# Serializing Data to JSON

* Python Inheritance
* Method parameters with default values
* Calling the inherited constructor logic
* Python's class member dictionary

```python
import numpy as np
import json

# =======================================================================================================================
class CSample(object):
  # --------------------------------------------------------------------------------------------------------
  def __init__(self, p_sJSON=None):
      #.......................... |  Instance Attributes | ............................
      #................................................................................
      if p_sJSON is not None:
        self.__dict__ = json.loads(p_sJSON)
  # --------------------------------------------------------------------------------------------------------
  def ToJSON(self):
    return json.dumps(self, default=lambda o: o.__dict__, sort_keys=True, indent=4)
  # --------------------------------------------------------------------------------------------------------  
# =======================================================================================================================
class CPoint3D(CSample):
    def __init__(self, p_sJSON=None):
      super(CPoint3D, self).__init__(p_sJSON)
      #.......................... |  Instance Attributes | ............................
      self.ID = None
      self.Target = None
      self.Features = [None]*3
      #................................................................................

# =======================================================================================================================


oPoint = CPoint3D()
oPoint.ID = 1
oPoint.Target = 0
oPoint.Features[:] = [3, 4, 5]

sJSONText = oPoint.ToJSON()
```

# Loading Numpy Arrays from JSON Data
* Converting a Python list to a numpy array

```python
# =======================================================================================================================
class CVector(CSample):
    def __init__(self, p_sJSON=None):
      super(CVector, self).__init__(p_sJSON)
      #.......................... |  Instance Attributes | ............................
      self.FeaturesVector = np.asarray(self.Features, np.float32)
      #................................................................................

# =======================================================================================================================


oPoint = CVector(sJSONText)
print("Data type:", type(oPoint.Features))
print("Numpy Data type:", oPoint.FeaturesVector.dtype)
print("Feature Vector:", oPoint.FeaturesVector)
```
